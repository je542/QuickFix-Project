# fake news dertection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = {
"text":[
"Breaking news: Scientists discover new planet",
"Government announces new economic plan",
"Aliens landed in New York yesterday",
"Celebrity reveals secret to immortality",
"New technology improves medical treatment",
"Fake cure for cancer discovered online",
"Experts warn about climate change",
"Man claims he can live without food",
"New education reforms announced",
"Secret society controls the world",
"New research improves cancer treatment",
"Government opens new hospitals",
"Fake diet pill spreads online",
"Scientists warn about fake medical news",
"Aliens seen flying over city",
"New AI technology announced",
"Miracle cure discovered on social media",
"Experts confirm climate crisis",
"Man claims to predict future",
"New vaccine successfully tested"
],

"label":[
"Real","Real","Fake","Fake","Real","Fake","Real","Fake","Real","Fake",
"Real","Real","Fake","Fake","Fake","Real","Fake","Real","Fake","Real"
]
}

df = pd.DataFrame(data)
df.head()
```

```python
X = df["text"]
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=42
)
```

```python
vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()

model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
news = ["Scientists confirm water on Mars"]

prediction = model.predict(vectorizer.transform(news))

print("Prediction:", prediction[0])
```
